```python
# passando colunas para minúsculo
df_credito.columns = [col.lower() for col in df_credito.columns]

data_path = df_credito
df_raw = ge.from_pandas(data_path)
```

```python
# Colunas requeridas, conforme dataframe novo
required_cols = [
    xxxxxxx
]

# Colunas não nulas, conforme dataframe novo
not_be_null_cols = [
    xxxxxxxxxxxxxx
]

# Data
date_col_name = [
    xxxxxxxxxx
]
date_col_format = '%Y-%m-%d'

# tipos de dados
type_cols = {
    qqqqqqqqqq
}

cols_above_zero = [
 qqqqqqqqqqqqq
]
```

```python
## Criando expectativas
# Conferindo existência de colunas
for col in required_cols:
        df_raw.expect_column_to_exist(column=col)

# Conferindo valores não nulos em colunas
if not_be_null_cols is not None:
        for col_name in not_be_null_cols:
            df_raw.expect_column_values_to_not_be_null(col_name, mostly=0.9)

# validando data
# if (date_col_name is not None) and (date_col_format is not None):
#          df_raw.expect_column_values_to_match_strftime_format(
#             date_col_name, date_col_format, mostly=0.9
#        )

# Conferindo tipos de dados nas colunas
for col_name, col_type in type_cols.items():
        df_raw.expect_column_values_to_be_of_type(col_name, col_type)

# Conferindo colunas com valores maiores que 0
for col in cols_above_zero:
         df_raw.expect_column_values_to_be_between(col, min_value=0, max_value=np.inf)

# # Checando colunas únicas
# for col_name in unique_cols:
#     df_raw.expect_column_values_to_be_unique(col_name)


## Validando resultados
valid_result = df_raw.validate()
if valid_result["success"] is True:
    status = "True"
else:
    status = "False"

# Printando status
status
```

```python
valid_report = ge.render.renderer.ValidationResultsPageRenderer().render(
        valid_result
    )

with open(data_validation_report, "w", encoding="utf-8") as writer:
        writer.write(ge.render.view.DefaultJinjaPageView().render(valid_report))
```
